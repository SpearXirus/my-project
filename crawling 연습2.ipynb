### 서울 소방서 위치 지도에 표시

```python
driver.get(url)
time.sleep(5)
driver.find_element(By.ID, 'searchKeyword').send_keys('서울')
driver.find_element(By.ID, 'fsSearchBtn').click()
soup = BeautifulSoup(driver.page_source, 'html.parser')
lis = soup.select('.stations-list > li')
```

```python
strong = soup.select('.stations-local-after > p > strong')
count = int(strong[1].get_text().replace('건','').strip())
count = math.ceil(count/10)
count
```

```python
lines = []
for i in range(count):
    if i >= 1 and i % 2 == 0:
        driver.find_element(By.CSS_SELECTOR,'.next_page').click()
        time.sleep(3)
    if i >= 1 and i % 2 == 1:
        driver.find_element(By.XPATH, '//*[@id="listForm"]/div/section/ul/li[1]/div/div/ul/li[4]/a').click()
        time.sleep(3)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    lis = soup.select('.stations-list > li')
    for li in lis:
        name = li.select_one('.title').get_text().strip()
        address = li.find('address').string.strip()
        tel = li.select_one('.tel').get_text().strip()
        lines.append([name, address, tel])
lines
```

```python
df = pd.DataFrame(lines, columns=('소방서','주소','전화번호'))
df
```

```python
kakao_api = 'c6fb0ac29226c99411f33661deb221d4'
uidos=[]
gyeongdoes=[]
for i in range(100):
    addr = df.iloc[i,1]
    search_url = 'https://dapi.kakao.com/v2/local/search/address.json'
    url2 = f'{search_url}?query={quote(addr)}'
    result2 = requests.get(url2,
                     headers={'Authorization': f'KakaoAK {kakao_api}'}).json()
    try:
        uidos.append(result2['documents'][0]['address']['y'])
    except:
        uidos.append('-')
    try:
        gyeongdoes.append(result2['documents'][0]['address']['x'])
    except:
        gyeongdoes.append('-')
print(uidos)
print(gyeongdoes)
```

```python
df['위도']=uidos
df['경도']=gyeongdoes
df
```

```python
seoul_map = folium.Map(location=[37.55,126.96],zoom_start=12)
for name, lat, lng in zip(df.소방서,df.위도,df.경도):
    try:
        folium.Marker([lat,lng],popup=name).add_to(seoul_map)
    except:
        pass
seoul_map
```

```python
df.to_csv('서울소방서.csv',index=False)
```
