```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import quote
import requests
import folium
```

```python
driver = webdriver.Chrome()
url = 'https://www.police.go.kr/user/bbs/BD_selectBbsList.do?q_bbsCode=1038&q_tab=1'
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
soup
```

```python
table = soup.select('tbody > tr')
table
```

```python
for i in table:
    for j in i.select('td'):
        print(j.text)
```

```python
for i in table:
    print(i.select('td')[0].text)
```

```python
office = []
address = []
homepage = []
for i in table:
    office.append(i.select('td')[0].text)
    address.append(i.select('td')[1].text)
    homepage.append(i.select('td')[2].text.strip('\n'))
print(office)
print(address)
print(homepage)
```

```python
dict = {'경찰서':office, '주소':address, '홈페이지':homepage}

df = pd.DataFrame(data=dict)
print(df)
```

```python
df.to_csv('서울 경찰서', index=False)
```

```python
kakao_api = 'c6fb0ac29226c99411f33661deb221d4'
uidos=[]
gyeongdoes=[]
for i in range(32):
    addr = df.iloc[i,1]
    search_url = 'https://dapi.kakao.com/v2/local/search/address.json'
    url2 = f'{search_url}?query={quote(addr)}'
    result2 = requests.get(url2,
                     headers={'Authorization': f'KakaoAK {kakao_api}'}).json()
    try:
        uidos.append(result2['documents'][0]['address']['y'])
    except:
        uidos.append('-')
    try:
        gyeongdoes.append(result2['documents'][0]['address']['x'])
    except:
        gyeongdoes.append('-')
print(uidos)
print(gyeongdoes)
```

```python
df['위도']=uidos
df['경도']=gyeongdoes
df
```

```python
seoul_map = folium.Map(location=[37.55,126.96],zoom_start=12)
for name, lat, lng in zip(df.경찰서,df.위도,df.경도):
    try:
        folium.Marker([lat,lng],popup=name).add_to(seoul_map)
    except:
        pass
seoul_map
```
